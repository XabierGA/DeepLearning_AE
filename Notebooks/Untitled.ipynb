```python
import pandas as pd
import numpy as np
```

```python
df_r = pd.read_csv("../Data/reactants_descriptors.csv" , index_col=0)
df_p = pd.read_csv("../Data/products_descriptors.csv" , index_col=0)
```

```python
desc = df_r.columns[1:-1]
```

```python
diff = df_r[desc]*df_p[desc]
diff["diff_ae"] = df_r["diff_ae"]
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("darkgrid")
g = sns.pairplot(diff , y_vars=np.array(desc), x_vars=["diff_ae"])
```
